```python
from utils import get_data

X_train, X_test, y_train, y_test, feature_names, preprocess_data = get_data()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

parameters = {
    "n_neighbors": range(3, 50),
    "weights": ["uniform", "distance"],
}
gridsearch = GridSearchCV(KNeighborsClassifier(), parameters)
gridsearch.fit(X_train, y_train)
```

```python
from utils import get_classification_metrics

knn_predictions = gridsearch.predict(X_test)
accuracy, report, matrix = get_classification_metrics(y_test, knn_predictions)

print(accuracy)
```
